```python
import json
import pandas as pd
import numpy as np
with open("ndis_providers.json", 'r', encoding="utf-8") as f:
    data = json.load(f)
    records = data.get("data", [])
df = pd.DataFrame(records)
df = df.head(100)
df['Prov_N'] = df['Prov_N'].str.replace(r'^[#!\s]+', '', regex=True).str.strip()
df['RegGroup'] = df['RegGroup'].astype(str)
df['RegGroup'] = df['RegGroup'].str.strip('[]')
#df['RegGroup'] = df['RegGroup'].apply(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df.fillna("N/A", inplace=True)
df = df[['Prov_N', 'ABN', 'Phone', 'Email', 'Website', 'Address', 'State_cd', 'Post_cd', 'RegGroup', 'prfsn']]
df.columns = ['Provider Name', 'ABN', 'Phone', 'Email', 'Website', 'Address', 'State Code', 'Post Code', 'Registration Group', 'Profession']
# rename_dict = {
#     "Prov_N": "Provider Name",
#     "ABN": "ABN",
#     "Head_Office": "Head Office",
#     "Outletname": "Outlet Name",
#     "Phone": "Phone",
#     "Website": "Website",
#     "Email": "Email",
#     "Address": "Address",
#     "State_cd": "State",
#     "Post_cd": "Postcode",
#     "RegGroup": "Registration Groups",
#     "opnhrs": "Opening Hours",
#     "prfsn": "Profession"
# }
# columns_to_keep = list(rename_dict.keys())
# df_final = df[columns_to_keep].rename(columns=rename_dict)
df.to_csv("ndis_providers.csv", index=False, encoding="utf-8-sig")
df.to_excel("ndis_providers.xlsx", index=False)
```
